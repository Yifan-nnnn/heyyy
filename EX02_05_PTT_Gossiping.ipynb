```python
import csv
articles = []
def get_articals(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "lxml")
  for atcl in soup.find_all("div", class_="r-ent"):
    if atcl.find("div", class_="title").a:
      title = atcl.find("div", class_="title").a.text
      href = "https://www.ptt.cc" + atcl.find("div", class_="title").a.get("href")
    else:
      title = atcl.find("div", class_="title").text
    date = atcl.find("div", class_="date").text
    articles.append([title, href, date])
    print(title, href, date)
  url = "https://www.ptt.cc" + soup.find_all("a", class_="btn wide")[1].get("href")
  return url
url = "https://www.ptt.cc/bbs/Gossiping/index.html"
article = [["Title", "URL", "Date"]]
for i in range(5):
  url = get_articals(url)
  article.extend(articles)
  print(article)

with open("article.csv", "w", newline="") as f:
  writer = csv.writer(f)
  writer.writerows(article)
```
